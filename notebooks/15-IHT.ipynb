```python
import sys
sys.path.insert(1, '../')
```

```python
import joblib
import pickle

from functions.data import load_dataset, DATASETS_LS
from configs.classical_models import estimator_dict
from configs.hyperparams import hyperparam_classical_dict
from functions.cv_undersamplers import undersample_data, train_model_w_undersampling
from functions.evaluation import evaluate_model_on_test_set

from imblearn.under_sampling import InstanceHardnessThreshold
```

```python
import warnings
warnings.filterwarnings("ignore", message="X does not have valid feature names")
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
scores_dict = {}

for dataset in DATASETS_LS:

    print(dataset)

    scores_dict[dataset] = {}

    # load dataset
    X_train, X_test, y_train, y_test = load_dataset(dataset)
          
    # train model with random search
    for estimator, params in zip(estimator_dict, hyperparam_classical_dict):

        # we define the undersampler with the model
        iht = InstanceHardnessThreshold(
            estimator=estimator_dict[estimator],
            random_state=0,
            cv=3,
        )

        # undersample data
        xtrainu, ytrainu, xtest, ytest, Xu, yu = undersample_data(iht, X_train, y_train)
    
        # check degree of undersampling:
        print(f"original data size: {len(X_train)/3*2}")
        print(f"undersampeld data: {len(xtrainu[0])}")
        print()
        
        # train model
        search = train_model_w_undersampling(
            estimator_dict[estimator], 
            hyperparam_classical_dict[params], 
            xtrainu, ytrainu, xtest, ytest, Xu, yu,
        )
        
        joblib.dump(search, f"../models/iht/{dataset}_{estimator}_iht.pkl")
        
        # evaluate model
        scores_dict[dataset][f"{estimator}_iht"] = evaluate_model_on_test_set(
            search, X_test, y_test)

# save intermediate results
with open("../models/iht/results", "wb") as fp:
    pickle.dump(scores_dict, fp)
```
