## Train classifiers

```python
# to store the results
scores_dict = {}

for dataset in DATASETS_LS:

    print(dataset)

    # initiate a dictionary per dataset
    scores_dict[dataset] = {}

    # load dataset
    X_train, X_test, y_train, y_test = load_dataset(dataset)

    # train models with hyperparameter tuning
    for estimator, params in zip(estimator_dict, hyperparam_classical_dict):

      search = train_model(
          estimator_dict[estimator], 
          hyperparam_classical_dict[params],
          X_train, 
          y_train,
      )

      # save model
      joblib.dump(search, f"../models/classical/{dataset}_{estimator}.pkl")

      # evaluate model
      scores_dict[dataset][estimator] = evaluate_model_on_test_set(search, X_test, y_test)


# save evaluation results
with open("../models/classical/results", "wb") as fp:
    pickle.dump(scores_dict, fp)
```
